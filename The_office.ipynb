##SEASONS

```python
seasons
```

```python
#en la tabla tenemos un multiindice no deseado. Eliminamos el primer nivel del índice de las columnas, ya que en la mayoría de los casos está repetido
#Y en el único caso en el que no está repetido (Originally aired), pensamos que no aporta demasiado a nivel explicativo, y que el df se entiende perfectamente sin este título
#añadimos el parámetro drop=True para que no se cree un índice de más
di_seasons = seasons.droplevel(level=0, axis=1).reset_index(drop=True)
di_seasons
```

```python
#hay una columna que sobra, la que tiene como título Episodes.1, ya que contiene datos repetidos. Procedemos a borrarla
di_seasons.drop(["Episodes.1"], axis = 1, inplace=True)
di_seasons
```

```python
#en la columna Rank aparecen unas notas que quiero eliminar
di_seasons['Rank'] = di_seasons['Rank'].str.replace(r'\[.*\]', '', regex=True)
di_seasons
```

```python
#vamos a comprobar el tipo de dato de cada columna
di_seasons.dtypes
```

```python
#pensamos que las columnas First aired y Last aired deberían de estar en formato fecha
di_seasons['First aired'] = pd.to_datetime(di_seasons['First aired'])
di_seasons['Last aired'] = pd.to_datetime(di_seasons['Last aired'])
```

```python
#también vemos más conveniente que la columna Rank sea de tipo entero
di_seasons['Rank'] = di_seasons['Rank'].astype(int)
```

```python
#volvemos a comprobar el tipo de dato de cada columna para verificar que todo está como nosotros deseamos
di_seasons.dtypes
```

```python
#pese a que la tabla es pequeña y se puede apreciar a simple vista, comprobamos si la tabla tiene nulos, cerciorándonos de que no tiene
di_seasons.isnull().sum()
```

```python
#Este es el aspecto definitivo de la tabla
di_seasons
```

En la tabla "characters" ya vimos que existen lo que podríamos llamar como etiquetas de protagonismo (Main, Starring, etc), que indican el peso del personaje en la temporada. Vamos a realizar un recuento de aquellos.

```python
#lo convertimos en una serie, concatenando las columnas, para poder hacer así un recuento de manera más sencilla
seasons_concat = pd.concat([only_seasons[col] for col in only_seasons.columns])
protagonism_count = seasons_concat.value_counts()
protagonism_count
```

El valor predominante es el símbolo "--", esto es, el valor que denota la ausencia de un personaje en una temporada. Luego, en orden de frecuencia, tenemos Starring, Recurring, Main, y Guest.

Recordamos que estos datos, por ejemplo, no indican que haya 39 personajes que hayan sido "Guest" en la serie. Cada personaje tiene una de estas etiquetas de protagonismo por cada temporada en la que haya aparecido; hay que tener en cuenta estas repeticiones a la hora de interpretar los resultados.

Vamos a establecer un sistema de puntuación de protagonismo para los personajes del show. Asignaremos una puntuación a cada tipo de protagonismo: por ejemplo, Main es un 5, y Starring un 4. Si un personaje no aparece en una temporada, obtendrá una puntuación de 0 en dicha. Se contabilizarán todos los personajes en todas las temporadas.

Esta será un ranking alternativo al que ya hicimos con characters_rank, pero pensamos que este es más completo porque además tiene en cuenta las etiquetas de protagonismo.

```python
#empezamos creando una función que asigne a cada personaje su puntuación de protagonismo
def score_characters(row):
    total_score = 0
    for season in row[2:11]:
        if season == 'Main':
            total_score += 5
        elif season == 'Starring':
            total_score += 4
        elif season == 'Recurring':
            total_score += 3
        elif season == 'Guest':
            total_score += 2
        else:
            total_score += 0
    return total_score
```

```python
#creamos una nueva columna en el df characters que contiene los resultados obtenidos por la función score_characters
characters['Score'] = characters.apply(score_characters, axis=1)
```

```python
#finalmente ya tenemos nuestra clasificación de personajes con su puntuación de protagonismo
characters_sc = characters[['Character', 'Score']].sort_values(by = 'Score', ascending = False).reset_index(drop = True)
characters_sc
```

Podemos ver cómo este ranking lo encabezan tres personajes: Jim Halpert, Pam Beesly y Dwight Schrute. Estos son los únicos personajes que aparecen como protagonistas en todas las temporadas de la serie.
