##SEASONS

```python
seasons
```

```python
#en la tabla tenemos un multiindice no deseado. Eliminamos el primer nivel del ndice de las columnas, ya que en la mayora de los casos est repetido
#Y en el nico caso en el que no est repetido (Originally aired), pensamos que no aporta demasiado a nivel explicativo, y que el df se entiende perfectamente sin este ttulo
#aadimos el parmetro drop=True para que no se cree un ndice de ms
di_seasons = seasons.droplevel(level=0, axis=1).reset_index(drop=True)
di_seasons
```

```python
#hay una columna que sobra, la que tiene como ttulo Episodes.1, ya que contiene datos repetidos. Procedemos a borrarla
di_seasons.drop(["Episodes.1"], axis = 1, inplace=True)
di_seasons
```

```python
#vamos a comprobar el tipo de dato de cada columna
di_seasons.dtypes
```

```python
#pensamos que las columnas First aired y Last aired deberan de estar en formato fecha
di_seasons['First aired'] = pd.to_datetime(di_seasons['First aired'])
di_seasons['Last aired'] = pd.to_datetime(di_seasons['Last aired'])
```

```python
#tambin vemos ms conveniente que la columna Rank sea de tipo entero
di_seasons['Rank'] = di_seasons['Rank'].astype(int)
```

```python
#volvemos a comprobar el tipo de dato de cada columna para verificar que todo est como nosotros deseamos
di_seasons.dtypes
```

```python
#pese a que la tabla es pequea y se puede apreciar a simple vista, comprobamos si la tabla tiene nulos, cerciorndonos de que no tiene
di_seasons.isnull().sum()
```

En la tabla "characters" ya vimos que existen lo que podramos llamar como etiquetas de protagonismo (Main, Starring, etc), que indican el peso del personaje en la temporada. Vamos a realizar un recuento de aquellos.

```python
#empezamos creando un nuevo dataframe solo con las columnas de las temporadas
columns_drop = ['Character', 'Portrayed by', 'Seasons_appeared']
only_seasons = characters.drop(columns=columns_drop)
```

```python
#lo convertimos en una serie, concatenando las columnas, para poder hacer as un recuento de manera ms sencilla
seasons_concat = pd.concat([only_seasons[col] for col in only_seasons.columns])
protagonism_count = seasons_concat.value_counts()
protagonism_count
```

El valor predominante es el smbolo "--", esto es, el valor que denota la ausencia de un personaje en una temporada. Luego, en orden de frecuencia, tenemos Starring, Recurring, Main, y Guest.

Recordamos que estos datos, por ejemplo, no indican que haya 39 personajes que hayan sido "Guest" en la serie. Cada personaje tiene una de estas etiquetas de protagonismo por cada temporada en la que haya aparecido; hay que tener en cuenta estas repeticiones a la hora de interpretar los resultados.

Vamos a establecer un sistema de puntuacin de protagonismo para los personajes del show. Asignaremos una puntuacin a cada tipo de protagonismo: por ejemplo, Main es un 5, y Starring un 4. Si un personaje no aparece en una temporada, obtendr una puntuacin de 0 en dicha. Se contabilizarn todos los personajes en todas las temporadas.

Esta ser un ranking alternativo al que ya hicimos con characters_rank, pero pensamos que este es ms completo porque adems tiene en cuenta las etiquetas de protagonismo.

```python
#empezamos creando una funcin que asigne a cada personaje su puntuacin de protagonismo
def score_characters(row):
    total_score = 0
    for season in row[2:11]:
        if season == 'Main':
            total_score += 5
        elif season == 'Starring':
            total_score += 4
        elif season == 'Recurring':
            total_score += 3
        elif season == 'Guest':
            total_score += 2
        else:
            total_score += 0
    return total_score
```

```python
#creamos una nueva columna en el df characters que contiene los resultados obtenidos por la funcin score_characters
characters['Score'] = characters.apply(score_characters, axis=1)
```

```python
#finalmente ya tenemos nuestra clasificacin de personajes con su puntuacin de protagonismo
characters_sc = characters[['Character', 'Score']].sort_values(by = 'Score', ascending = False).reset_index(drop = True)
characters_sc
```

Podemos ver cmo este ranking lo encabezan tres personajes: Jim Halpert, Pam Beesly y Dwight Schrute. Estos son los nicos personajes que aparecen como protagonistas en todas las temporadas de la serie.
